```python
%%time

# Trace evaluation domain and ZKSTARK evaluation domain
# T = trace evaluation domain
# L = ZKSTARK evaluation domain
# T and L are disjoint
# |T| < |L|
# |T| is a power of 2 (to use FFT)
# |L| is a power of 2 (to use FFT)
# F[P] = Prime field of order P, P is prime
# F[P, *] = Multiplicative subgroup of F[P] = {1, 2, 3, ..., P - 1}
# |F[P, *]| = P - 1
# T and L are subgroups of F[P, *] -> |T| and |L| divides |F[P, *]| = P - 1
import fft_poly
import iop
import merkle
import numpy as np
import polynomial
import stark
from fft import eval_poly, fft, ifft
from field import F, find_generator, generate, get_primitive_root
from polynomial import Polynomial, X

from utils import is_pow2, is_prime, min_pow2_gt

### Example claim = trace is all 0 or 1 ###
trace = [int(i) for i in np.random.choice([0, 1], size=1024)]
trace_len = len(trace)

### Setup for evaluation domains - pick prime field and a generator ###
# Prime field F[P]
# Starkware prime
# P = 2**251 + 17 * 2**192 + 1
P = 3 * 2**30 + 1
assert is_prime(P)

# Generator of F[P, *] - multiplicative group of the prime field F[P]
# g must be known by the prover and the verifier
g = find_generator(P)
print("Generator of F[P, *] = g =", g)

### Trace evaluation domain ###
# Let T = trace evaluation domain
# |T| must be a power of 2 to use FFT
# |T| must divide P - 1 so that |T| is a subgroup of F[P, *]
assert is_pow2(trace_len)
assert (P - 1) % trace_len == 0
trace_domain = generate(get_primitive_root(g, trace_len, P), trace_len, P)

### Trace polynomial ###
# t(x) = trace polynomial
# t(T[i]) = trace[i]
t = fft_poly.interp(trace, trace_domain, P)
assert fft_poly.eval(t, trace_domain, P) == trace, "t(T) != trace"
print("Degree of trace polynomial =", t.degree())

### Constraint polynomial ###
# c(x) = constraint polynomial
# trace[i] is either 0 or 1 -> trace[i](trace[i] - 1) = 0 for all T[i]
#                           -> t(x)(t(x) - 1) = 0 for all T[i]
#                           -> Set c(x) = t(x)(t(x) - 1)
c = t * t - t
print("Degree of constraint polynomial =", c.degree())

EXP_FACTOR = 8
N = EXP_FACTOR * trace_len

prover = stark.Prover(
    P=P,
    g=g,
    trace_poly=t,
    trace_domain=trace_domain,
    exp_factor=EXP_FACTOR,
    constraint_poly=c
)

x = X(1, lambda x: F(x, P))

verifier = stark.Verifier(
    P=P,
    g=g,
    trace_len=trace_len,
    exp_factor=EXP_FACTOR,
    constraint_poly=x * x - x
)

### Communication channels ###
chan = iop.Channel(iop.Prover(prover), iop.Verifier(verifier))

### STARK and FRI - prover commit ###
prover.commit(chan)

### STARK - verifier query ###
# Preliminary check before queries
verifier.check()

NUM_QUERIES = 50
assert NUM_QUERIES <= N

Q = [int(i) for i in np.random.choice(range(N), size=NUM_QUERIES, replace=False)]
for i in Q:
    verifier.query(i, chan)

### FRI - verifier query ###
for i in Q:
    verifier.fri().query(i, chan)

print("All checks passed ✅️")

# TODO - attacks
# TODO - probabilities?
# TODO: pair t(x) and t(-x) in merkle tree?
```
