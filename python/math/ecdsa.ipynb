```python
from hashlib import sha256
from ecdsa import SECP256k1
from ecdsa.ecdsa import Private_key, Public_key, generator_secp256k1

# Curve parameters
G = generator_secp256k1
n = G.order()

x = 0x123456789ABCDEF123456789ABCDEF123456789ABCDEF123456789ABCDEF
pub_key = Public_key(G, x * G)
priv_key = Private_key(pub_key, x)

# Sign 2 different messages with the same nonce
k = 123456789012345678901234567890123456789

m1 = b"Hello"
m2 = b"World"

z1 = int.from_bytes(sha256(m1).digest(), "big")
z2 = int.from_bytes(sha256(m2).digest(), "big")

s1 = priv_key.sign(z1, k)
s2 = priv_key.sign(z2, k)

print("Private key")
print(hex(x))
print("Nonce")
print(hex(k))
print()


# Attacker recovers private key

# Modular inverse
def inv(x):
    return pow(x % n, -1, n)

rec_k = ((z1 - z2) * inv(s1.s - s2.s)) % n
rec_priv = ((rec_k * s1.s - z1) * inv(s1.r)) % n

print("Recovered nonce")
print(hex(rec_k))
print("Recovered private key")
print(hex(rec_priv))
print()

assert(rec_priv == x)
```
