```python
# Euclidean algo
# gcd(a, b) = gcd(b, a mod b)

# gcd(48, 18) -> gcd(18, 12)
#             -> gcd(12, 6)
#             -> gcd(6, 0)

def gcd(a: int, b: int) -> int:
    while b != 0:
        a, b = b, a % b
    return a

gcd(48, 18)
```

```python
# Extended Euclidean algo -> finds gcd(a, b) = sa + tb
# for a < p, p prime
# gcd(a, p) = 1 -> sa + tp = 1 -> sa = 1 mod p -> s = a^-1

def egcd(a, b) -> (int, int, int):
    r0, r1 = a, b
    s0, s1 = 1, 0
    t0, t1 = 0, 1

    while r1 != 0:
        q = r0 // r1
        r0, r1 = r1, r0 - q * r1
        s0, s1 = s1, s0 - q * s1
        t0, t1 = t1, t0 - q * t1

    # gcd(a, b), s, t
    return r0, s0, t0

egcd(3, 11)
```
