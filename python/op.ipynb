```python
# Trace ancestor

# index of most significant bit
def msb(x):
    i = 0
    x >>= 1
    while x != 0:
        i += 1
        x >>= 1
    return i

# set 1 for right most unset bit
def lsb(x):
    return ~x & (x + 1)

def fmt(x):
    return format(x, "06b")

print("i fmt(i) fmt(l) m a")
for i in range(16, 32):
    l = lsb(i)
    m = msb(l)
    a = i >> m
    print(i, fmt(i), fmt(l), m, f'{a: >2}')    
```
